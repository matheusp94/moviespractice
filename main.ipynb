# Transformando colunas em listas

    - read_csv -> possibilita armazenar os dados em "data" a partir de um DataFrame criado pelo pandas.
    - abreviação -> data['nome_coluna'].tolist() extrai valores da de uma coluna do DataFrame convertendo para uma lista nome_coluna: []

 

```python
# Convertendo cada coluna do csv em uma lista

from pandas import *
 
# reading CSV file
data = read_csv("movies.csv")
 
# converting column data to list
tt = data['title'].tolist()
yr = data['year'].tolist()
rt = data['runtime'].tolist()
ct = data['certificate'].tolist()
gn = data['genre'].tolist()
dr = data['director'].tolist()
st = data['stars'].tolist()
ra = data['rating'].tolist()
ms = data['metascore'].tolist()
vt = data['votes'].tolist()

 
''' usar para verificar se as listas estão formatadas corretamente
print('title:', tt)
print('year:', yr)
print('runtime:', rt)
print('certificate:', ct)
print('director:', dr)
print('stars:', st)
print('rating:', ra)
print('metascore:', ms) 
print('votes:', vt) '''
```

# Variáveis numéricas do conjunto de dados
    - runtime:
        Duração do filme em minutos
    - rating: 
        Classificação do filme por usuários registrados na IMDB. Varia de 1 a 10
    - metascore:
        Média ponderada de muitas avaliações provenientes de críticos de renome. A equipe do Metacritic lê as resenhas e atribui a cada uma uma pontuação de 0 a 100, que recebe um peso, principalmente com base na qualidade e na fonte da resenha.
    - votes:
        Quantidades de votos feitos por usuários registrados na IMDB.

```python
mh_metascore = len(ms_tratado) / np.sum(1 / np.array(ms_tratado))
print(f'A média harmônica de classificação dentre todos os filmes listados é de {mh_metascore} pontos.')
```
